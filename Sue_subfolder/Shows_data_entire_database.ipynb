```python
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
url='http://api.tvmaze.com/shows?page='
```

```python
page_list=np.arange(165)
page_list
```

```python
id_=[]
name=[]
url_=[]
language=[]
genre=[]
premiered=[]
rating=[]
country=[]
image=[]
summary=[]
updated=[]
error_list=[]

for n in page_list:
    response = requests.get(url+str(page_list[n])).json()
    print("Performing page "+str(n))
    for i in np.arange(len(response)):
        try:
            id_.append(int(response[i]["id"]))
        except: 
            id_.append("NaN")
        try:
            name.append(response[i]["name"])
        except:
            name.append("NaN")
        try:
            url_.append(response[i]["url"])
        except:
            url_.append("NaN")
        try:
            language.append(response[i]["language"])
        except: 
            language.append("NaN")
        try:
            genre.append(response[i]["genres"][0].encode(encoding='UTF-8',errors='strict'))
        except:
            genre.append("Nan")
        try:
            premiered.append(response[i]["premiered"].encode(encoding='UTF-8',errors='strict'))
        except:
            premiered.append("NaN")
        try:
            rating.append(response[i]["rating"]["average"])
        except:
            rating.append("NaN")
        try:
            country.append(response[i]["network"]["country"]["name"].encode(encoding='UTF-8',errors='strict'))
        except:
            country.append("NaN")
        try:
            image.append(response[i]["image"]["original"].encode(encoding='UTF-8',errors='strict'))
        except:
            image.append("NaN")
        try:
            summary.append(response[i]["summary"])
        except:
            summary.append("NaN")
        try:
            updated.append(response[i]["updated"].encode(encoding='UTF-8',errors='strict'))
        except:
            updated.append("NaN")
    
```
